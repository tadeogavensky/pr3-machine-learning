## K-Means — Ejercicio 4

**Objetivo:** calcular la distancia entre dos puntos.

**Input:**

```python
p1 = (1, 1)
p2 = (2, 1)
```

**Output esperado:**

```python
1.0
```

Pista conceptual: distancia euclidiana.



````python
import numpy as np
import math

p1 = (1, 1)
p2 = (2, 1)


def calc_euclidian_dis_np(p1, p2):
    v1, v2 = np.array(p1), np.array(p2)

    return np.sqrt(np.sum((v1 - v2) ** 2))


def calc_euclidian_dis_py(p1, p2):

    return math.sqrt(((p1[0] - p2[0]) ** 2) + ((p1[1] - p2[1]) ** 2))


print("Numpy", calc_euclidian_dis_np(p1, p2))
print("Python", calc_euclidian_dis_py(p1, p2))
````

## K-Means — Ejercicio 5

**Objetivo:** calcular la distancia entre `(1, 1)` y `(4, 5)`.

**Output esperado aproximado:**

```python
5.0
```

````python
p1 = (1, 1)
p2 = (4, 5)


def calc_euclidian_dis(p1, p2):
    v1, v2 = (np.array(p1), np.array(p2))

    return np.sqrt(np.sum((v1 - v2) ** 2))


print("Distance:", calc_euclidian_dis(p1, p2))
````

## K-Means — Ejercicio 6

**Objetivo:** calcular el centroide/promedio de todos los puntos.

**Input:**

```python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
```

**Output esperado:**

```python
(3.0, 2.25)
```

````python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]


def calc_centroid(X):

    mean_x = sum(x for x, y in X) / len(X)
    mean_y = sum(y for x, y in X) / len(X)

    return (mean_x, mean_y)

print("Centroid:", calc_centroid(X))
````

## K-Means — Ejercicio 9

**Objetivo:** calcular la distancia desde un punto a cada centroide.

**Input:**

```python
punto = (2, 1)
centroides = [(1, 1), (5, 4)]
```

**Output esperado aproximado:**

```python
[1.0, 4.24]
```

````python
point = (2, 1)
centroids = [(1, 1), (5, 4)]


def calc_distance_to_cetroids(point, centroids):

    distances = []
    for c in centroids:
        distances.append(calc_euclidian_dis(point, c))

    return distances


print(
    f"Distance from point {point} to cetroids {centroids}: \n",
    calc_distance_to_cetroids(point, centroids),
)
````

## K-Means — Ejercicio 10

**Objetivo:** decidir cuál es el centroide más cercano para el punto `(2, 1)`.

**Output esperado:**

```python
0
```

Interpretación: el punto pertenece al cluster del centroide en posición 0.

````python
point = (2, 1)


def closest_centroid(point, centroids):
    print("Centroids:", centroids)
    distances = calc_distance_to_cetroids(point, centroids)
    min_centroid = np.argsort(distances)[0]
    return min_centroid


print(f"Centroid index to point {point}: ", closest_centroid(point, centroids))
````

## K-Means — Ejercicio 11

**Objetivo:** asignar todos los puntos a su centroide más cercano.

**Input:**

```python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
centroides = [(1, 1), (5, 4)]
```

**Output esperado:**

```python
[0, 0, 1, 1]
```

````python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
centroids = [(1, 1), (5, 4)]


def assign_point_to_centroids(X, centroids):
    assigned_centroids = []

    for point in X:
        distances = []
        for c in centroids:
            distances.append(calc_euclidian_dis(point, c))

        min_distance_index = np.argsort(distances)[0]

        assigned_centroids.append(min_distance_index)

    return assigned_centroids


print(f"Assigned centroids to points - {X}: {assign_point_to_centroids(X,centroids)}")
````

## K-Means — Ejercicio 12

**Objetivo:** separar puntos por cluster usando labels.

**Input:**

```python

labels = [0, 0, 1, 1]
```

**Output esperado:**

Cluster 0:

```python
[(1, 1), (2, 1)]
```

Cluster 1:

```python
[(4, 3), (5, 4)]
```

````python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
labels = [0, 0, 1, 1]


def split_points_in_clusters(X, labels):

    cluster_0 = []
    cluster_1 = []

    for point, label in zip(X, labels):
        print(f"Point {point}, label {label}")
       
        if label == 0:
            cluster_0.append(point)
        else:
             cluster_1.append(point)

    return cluster_0, cluster_1


split_points_in_clusters(X, labels)
````
