## K-Means — Ejercicio 4

**Objetivo:** calcular la distancia entre dos puntos.

**Input:**

```python
p1 = (1, 1)
p2 = (2, 1)
```

**Output esperado:**

```python
1.0
```

Pista conceptual: distancia euclidiana.



````python
import numpy as np
import math

p1 = (1, 1)
p2 = (2, 1)


def calc_euclidian_dis_np(p1, p2):
    v1, v2 = np.array(p1), np.array(p2)

    return np.sqrt(np.sum((v1 - v2) ** 2))


def calc_euclidian_dis_py(p1, p2):

    return math.sqrt(((p1[0] - p2[0]) ** 2) + ((p1[1] - p2[1]) ** 2))


print("Numpy", calc_euclidian_dis_np(p1, p2))
print("Python", calc_euclidian_dis_py(p1, p2))
````

Interpretación: el punto pertenece al cluster del centroide en posición 0.

````python
point = (2, 1)


def closest_centroid(point, centroids):
    print("Centroids:", centroids)
    distances = calc_distance_to_cetroids(point, centroids)
    min_centroid = np.argsort(distances)[0]
    return min_centroid


print(f"Centroid index to point {point}: ", closest_centroid(point, centroids))
````

## K-Means — Ejercicio 16

**Objetivo:** ejecutar una iteración completa.

Debe hacer:

1. asignar labels
2. recomputar centroides

**Input inicial:**

```python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
centroides = [(1, 1), (5, 4)]
```

**Output esperado:**

```python
labels = [0, 0, 1, 1]
nuevos_centroides = [(1.5, 1.0), (4.5, 3.5)]
```

````python
points = [(1, 1), (2, 1), (4, 3), (5, 4)]
centroids = [(1, 1), (5, 4)]


def k_means(points, centroids):
    # Como asigno clusters (label) a unos puntos?
    # Recorro los puntos y los centroides
    # Por cada punto pregunto cual es su centroide mas cercano
    # El indice dentro de la lista de centroides es el label del cluster
    cluster_labels = []
    for p in points:
        # calculo distancias de punto actual a los centroides
        distances = []
        assigned_label = None
        for c in centroids:
            distances.append(
                calc_euclidian_dis(p, c)
            )  # ya tengo las distancias de (1, 1) a [(1, 1), (5, 4)]
            # cual es la mas cercana?

        closest_distance_index = np.argsort(distances)[
            0
        ]  # indice de la distancia mas cercana en la lista de distancias

        assigned_centroid = centroids[closest_distance_index]
        assigned_label = closest_distance_index

        print(
            f"El punto {p} fue asignado al centroide -> {assigned_centroid} y al cluster -> {assigned_label}"
        )
        cluster_labels.append({"point": p, "assigned_label": assigned_label})

    # recomputar centroides

    cluster_0_points = [
        cluster["point"] for cluster in cluster_labels if cluster["assigned_label"] == 0
    ]
    cluster_1_point = [
        cluster["point"] for cluster in cluster_labels if cluster["assigned_label"] == 1
    ]

    cluster_0_centroid = (
        sum(x for x, y in cluster_0_points) / len(cluster_0_points),
        sum(y for x, y in cluster_0_points) / len(cluster_0_points),
    )

    cluster_1_centroid = (
        sum(x for x, y in cluster_1_point) / len(cluster_1_point),
        sum(y for x, y in cluster_1_point) / len(cluster_1_point),
    )

    return (
        cluster_labels,
        [cluster_0_centroid, cluster_1_centroid],
    )


labels, new_centroids = k_means(points, centroids)

print(labels)
print(new_centroids)
````

## K-Means — Ejercicio 17

**Objetivo:** repetir otra iteración con los nuevos centroides.

**Output esperado:**

```python
labels = [0, 0, 1, 1]
```

Pregunta: ¿cambiaron las labels? ¿Qué significa eso?

````python
points = [(1, 1), (2, 1), (4, 3), (5, 4)]
centroids = (1.5, 1.0), (4.5, 3.5)


def k_means(points, centroids):
    labels = []
    for p in points:
        distances = []
        for c in centroids:
            distances.append(calc_euclidian_dis(p, c))

        min_distance_index = np.argsort(distances)[0]

        labels.append(min_distance_index)

    cluster_0 = []
    cluster_1 = []

    for point, label in zip(points, labels):
        if label == 0:
            cluster_0.append(point)
        else:
            cluster_1.append(point)

    recomputed_centroid_0 = (
        sum(x for x, y in cluster_0) / len(cluster_0),
        sum(y for x, y in cluster_0) / len(cluster_0),
    )
    recomputed_centroid_1 = (
        sum(x for x, y in cluster_1) / len(cluster_1),
        sum(y for x, y in cluster_1) / len(cluster_1),
    )

    return labels, (recomputed_centroid_0, recomputed_centroid_1)


k_means(points, centroids)

# No cambiaron los labels
# Se llego al punto de convergencia
````

## KNN — Ejercicio 10

**Objetivo:** crear una función manual que reciba:

- `X`
- `y`
- `nuevo_punto`
- `k`

y devuelva la predicción.

**Input de prueba:**

```python
nuevo_punto = (3, 2)
k = 3
```

**Output esperado:**

```python
'A'
```

````python
X = [(1, 1), (2, 1), (4, 3), (5, 4)]
y = ["A", "A", "B", "B"]
new_point = (3, 2)
k = 3


def knn_predict(X, y, new_point, k):
    points__with_labels = []

    for point, label in zip(X, y):
        points__with_labels.append((point, label))

    all_neighbors = []
    for p_label in points__with_labels:
        point = p_label[0]
        label = p_label[1]
        distance = calc_euclidian_dis(new_point, point)
        all_neighbors.append((distance, point, label))

    distances = [d for d, _, _ in all_neighbors]
    k_nn_positions = np.argsort(distances)[:k]

    k_nearest_neighbors = [all_neighbors[idx] for idx in k_nn_positions]
    labels = [label for _, _, label in k_nearest_neighbors]
    return Counter(labels).most_common(1)[0][0]


knn_predict(X, y, new_point, k)
````

## KNN — Ejercicio 11

**Objetivo conceptual:** comparar `k = 1` vs `k = 3`.

Respondé en Markdown:

- Qué cambia.
- Cuál es más sensible al ruido.
- Por qué KNN casi no “entrena” en comparación con otros algoritmos.

````python
# K = 1 toma únicamente el vecino más cercano al punto nuevo. Es más sensible al ruido porque, si ese vecino está mal etiquetado o es un caso raro, la predicción queda totalmente afectada.

# K = 3 toma los 3 vecinos más cercanos y decide por votación. Suele ser más estable porque no depende de un solo punto, aunque puede pasar que gane una clase por cantidad aunque uno de los vecinos esté un poco más cerca.

#KNN casi no “entrena” porque en fit solo guarda el dataset. No aprende pesos, reglas, centroides ni parámetros. Cuando llega un punto nuevo, calcula distancias contra los datos guardados y vota según los vecinos más cercanos.
````
